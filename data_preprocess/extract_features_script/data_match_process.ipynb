## Genearate gloss + frame_array dataset

```python
matched_samples = []
for entry in content:
    instances = entry['instances']
    gloss = entry['gloss']

    for inst in instances:
        video_id = inst['video_id']
        if video_id + '.mp4' in available_filenames:
            # print(f"video_id = {video_id}")
            matched_samples.append([video_id, gloss])

with open('matched_samples.txt', 'w') as f:
    f.write("video_file\tgloss\n")
    for video_file, gloss in matched_samples:
        print(f"video = {video_file}, gloss = {gloss}")
        f.write(f"{video_file}\t{gloss}\n")
```

```python
# matched_samples = pd.read_csv(r"D:\Courses\Pattern Recognition\UB-CSE-555\dataset\matched_samples.txt", sep='\t')

dataset = []
base_path = r"D:\Courses\Pattern Recognition\WLASL\WLASL\dataset\matrix"

for video, gloss in matched_samples:
# for row in matched_samples.itertuples(index=False):
#     gloss = row.gloss
#     video = row.video_file
    file_path = os.path.join(base_path, f"{video}.npy")
    print(file_path)
    if os.path.exists(file_path):
        matrix = np.load(os.path.join(base_path, f"{video}.npy"))
        dataset.append([gloss, matrix])
       
np.save("D:/Courses/Pattern Recognition/UB-CSE-555/data_preprocess/dataset/dataset_all_new.npy", np.array(dataset, dtype=object))
```

## Add paddings make the video array structure consistent

```python
dataset_padding = []
# First pass to find the maximum number of frames (T)
T_max = 0
# base_path = "./dataset/matrix"
base_path = r"D:\Courses\Pattern Recognition\WLASL\WLASL\dataset\matrix"

for video, _ in matched_samples:
    file_path = os.path.join(base_path, f"{video}.npy")
    if os.path.exists(file_path):
        matrix = np.load(file_path)
        if (T_max < matrix.shape[0]):
            print(file_path)
        T_max = max(T_max, matrix.shape[0])  # update max frame length

print(f"Maximum number of frames: {T_max}")

# Second pass: pad matrices to T_max
for video, gloss in matched_samples:
    file_path = os.path.join(base_path, f"{video}.npy")
    if os.path.exists(file_path):
        matrix = np.load(file_path)
        T, D = matrix.shape  # T: frames, D: features (200)
        
        # Pad with zeros if needed
        if T < T_max:
            pad_amount = T_max - T
            pad_matrix = np.zeros((pad_amount, D), dtype=np.float32)
            matrix = np.vstack([matrix, pad_matrix])  # [T_max, D]

        dataset_padding.append([gloss, matrix])

# Save as object array
np.save("D:/Courses/Pattern Recognition/UB-CSE-555/data_preprocess/dataset/dataset_all_padding_new.npy", np.array(dataset_padding, dtype=object))
```
