```python
import numpy as np
import matplotlib.pyplot as plt

# Load your video data
video_data = np.load(r"C:\Users\19692\Downloads\UB-CSE-555\st-gcn\final_dataset\videos_300_train.npy")  # or full path
sample = video_data[0]  # shape: (2, 64, 67, 1)

# Remove the last dimension
sample = sample.squeeze(-1)  # now shape: (2, 64, 67)

# Choose a frame index to visualize
frame_idx = 2
x = sample[0, frame_idx, 3]  # x-coordinates of all joints in that frame
y = sample[1, frame_idx, 3]  # y-coordinates

# Plot
plt.figure(figsize=(5, 5))
plt.scatter(x, -y, c='blue')  # Flip y for natural display
plt.title(f"Frame {frame_idx} of Sample 0")
plt.axis('equal')
plt.grid(True)
plt.show()
```
